```python
import datetime

import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
ds = pd.read_excel('locationFirst.xlsx', sheet_name='Reviews', decimal=",", engine='openpyxl')
ds.describe(include='all', datetime_is_numeric=True).T
```

```python
ds.dropna(subset=['userISO'], inplace=True)
ds.info()
```

```python
geolocator = Nominatim(user_agent="coordinates")

for i in ds.index:
    if len(ds['userISO'][i]) > 2:
        try:
            location = geolocator.geocode(ds['userISO'][i])
            ds['userISO'][i] = location.address
            print(ds['userISO'][i])
        except:
            print(ds['userISO'][i])
            pass
    else:
        print(ds['userISO'][i])
```

```python
ds.to_excel('locationSecond.xlsx', index=False, sheet_name='Reviews')

print(datetime.datetime.now())
```
